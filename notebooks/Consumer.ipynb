```python
# FAKHRI AZIS BASIRI
# ASSIGNMENT KAFKA

import json
from kafka import KafkaConsumer
from dotenv import load_dotenv
from pathlib import Path
import os

# Load environment variables
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)

# Kafka configuration
kafka_host = os.getenv('KAFKA_HOST')

# Initialize Kafka consumer
consumer = KafkaConsumer(
    'assignment-kafka',  # Topic name
    bootstrap_servers=f'{kafka_host}:9092',
    auto_offset_reset='earliest',
    value_deserializer=lambda m: json.loads(m.decode('utf-8'))
)

# Variables to hold running aggregation values
total_price = 0
total_quantity = 0
total_count = 0

print("Starting Consumer...")

for message in consumer:
    data = message.value
    price = data.get('price', 0)  # Get the price from the message; default to 0 if not found
    quantity = data.get('quantity', 0)  # Get the quantity from the message; default to 0 if not found

    # Update running sum and count
    total_price += price
    total_quantity += quantity
    total_count += 1

    # Calculate running averages
    running_average_price = total_price / total_count if total_count > 0 else 0
    running_average_quantity = total_quantity / total_count if total_count > 0 else 0

    print(f"Transaction ID: {data['transaction_id']}, Price: {price}, Quantity: {quantity}")
    print(f"Running Sum - Total Price: {total_price}, Total Quantity Sold: {total_quantity}")
    print(f"Running Average - Average Price: {running_average_price}, Average Quantity Sold: {running_average_quantity}\n")
```
